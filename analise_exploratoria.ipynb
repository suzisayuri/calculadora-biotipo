2. Configura√ß√£o do Ambiente
Abaixo, importamos as bibliotecas de an√°lise e carregamos o dataset gerado na etapa de Engenharia de Dados.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configura√ß√£o est√©tica
sns.set(style="whitegrid")

# 1. Carregar os dados (Simulando conex√£o com banco de dados)
try:
    df = pd.read_csv('dados_biotipos_simulados.csv')
    print("‚úÖ Dados carregados com sucesso!")
except FileNotFoundError:
    print("‚ùå Erro: Rode o Bloco 1 primeiro para gerar o arquivo CSV.")
```

## 3. An√°lise de Distribui√ß√£o dos Biotipos
Para definir o estoque e a grade de produ√ß√£o, precisamos entender qual √© o "perfil corporal" predominante no nosso p√∫blico.

A classifica√ß√£o dos biotipos foi feita com base na rela√ß√£o entre **Ombro, Busto, Cintura e Quadril**.

```python
# 2. Gr√°fico de Barras: Distribui√ß√£o dos Biotipos
plt.figure(figsize=(10, 6))
sns.countplot(x='biotipo', 
              hue='biotipo', 
              data=df, 
              palette='viridis', 
              order=df['biotipo'].value_counts().index,
              legend=False) 

plt.title('Distribui√ß√£o de Biotipos do P√∫blico (Simulado)', fontsize=16)
plt.ylabel('Quantidade de Pessoas')
plt.xlabel('Tipo de Corpo')
plt.show()
```

### üí° Insights Preliminares
* Observamos a predomin√¢ncia de certos biotipos, o que sugere onde devemos focar o desenvolvimento de novos moldes.
* A presen√ßa de biotipos variados valida a necessidade de uma grade de tamanhos flex√≠vel.

## 4. Correla√ß√£o: Cintura vs. Quadril
Agora, vamos investigar a rela√ß√£o matem√°tica entre a **Cintura** e o **Quadril**.

* **Linha Vermelha (1:1):** Representa uma propor√ß√£o "quadrada" (Cintura igual ao Quadril).
* **Pontos Acima da Linha:** Indicam corpos onde o quadril √© maior que a cintura (comum em Peras e Ampulhetas).
* **Pontos Abaixo da Linha:** Indicam cintura maior que quadril (comum em Ovais ou Tri√¢ngulos Invertidos).

```python
# 3. Gr√°fico de Dispers√£o: Cintura vs Quadril
plt.figure(figsize=(10, 6))
sns.scatterplot(x='cintura', y='quadril', hue='biotipo', data=df, alpha=0.6)

# Linha de refer√™ncia quadrada
plt.plot([60, 140], [60, 140], 'r--', linewidth=2, label='Linha 1:1 (Quadrado)')
```
